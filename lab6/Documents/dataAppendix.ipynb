```python
import pandas as pd
import matplotlib.pyplot as plt

DIR = "../AnalysisData/"

data = pd.read_csv(f"{DIR}brand_ratings.csv", sep=';')
print(data)
```

```python
yesSeries = data['ocena_K']
nodSeries = data['ocena_M']
marka = data['marka']

plt.figure(figsize=(30, 30))
df = pd.DataFrame({"kobiety":yesSeries,"mężczyźni":nodSeries, 'marka':marka})
ax = df.plot.bar(x='marka', color=["deeppink","deepskyblue"], rot=0, title="Średnie ocen "
                    "odkurzaczy danych marek\n z podziałem na kobiety i mężczyzn")
ax.set_xlabel("Marka")
ax.set_ylabel("Średnia ocena")
ax.grid(axis = 'y')
ax.set_ylim(0,3.5)

plt.show()
```

```python
yesSeries = data['sredni_wiek_K']
nodSeries = data['sredni_wiek_M']
marka = data['marka']

plt.figure(figsize=(30, 30))
df = pd.DataFrame({"kobiety":yesSeries,"mężczyźni":nodSeries, 'marka':marka})
ax = df.plot.bar(x='marka', color=["deeppink","deepskyblue"], rot=0, title="Średni wiek "
                    "kupujących odkurzacze danych marek\n z podziałem na kobiety i mężczyzn")
ax.set_xlabel("Marka")
ax.set_ylabel("Średni wiek w latach")
ax.grid(axis = 'y')
ax.set_ylim(0,55)

plt.show()
```

```python
yesSeries = data['dni_po_zakupie_K']
nodSeries = data['dni_po_zakupie_M']
marka = data['marka']

plt.figure(figsize=(30, 30))
df = pd.DataFrame({"kobiety":yesSeries,"mężczyźni":nodSeries, 'marka':marka})
ax = df.plot.bar(x='marka', color=["deeppink","deepskyblue"], rot=0, title="Średnia ilość "
                    "dni po zakupie odkurzacza w momencie składania oceny")
ax.set_xlabel("Marka")
ax.set_ylabel("Ilość dni po zakupie")
ax.grid(axis = 'y')
ax.set_ylim(0,10)

plt.show()
```

```python
data2 = pd.read_csv(f"{DIR}days_rating.csv", sep=';')

plt.figure(figsize=(30, 30))
ax = data2.plot.bar(x='dni_od_zakupu', y='ocena', title="Średnia ocena "
                    "w zależności od czasu po zakupie", label = "ocena", width=0.9, color='darkorange')
ax.set_xlabel("Ilość dni po zakupie")
ax.set_ylabel("Średnia ocena")
ax.grid(axis = 'y')
ax.set_ylim(1.75,3)

plt.show()
```
